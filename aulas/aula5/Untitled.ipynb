```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = (50.0, 40.0)
%matplotlib inline
```

```python
data = pd.read_csv("../../dados/historico_tempo_curado.csv", sep=";")
```

```python
data
```

```python
meses, anos = [], []
for d in data.Data:
    dd = d.split("/")
    print(dd)
    meses.append(dd[1])
    anos.append(dd[-1])
```

```python
import seaborn as sea
```

```python
temp_max = pd.DataFrame({"meses":meses, "anos":anos, "temp_max":data.TempMaximaMedia}).pivot("meses",
                                                                                             "anos",
                                                                                             "temp_max")
sea.set_context("poster")
f, ax = plt.subplots(figsize=(40, 10))
ax.set_title("Médias de temp. máximas")
with sea.color_palette("Reds_r") as colors:
    sea.heatmap(temp_max, annot=True,linewidths=.5, ax=ax, cmap=colors)
```

```python
temp_min = pd.DataFrame({"meses":meses, "anos":anos, "temp_min":data.TempMinimaMedia}).pivot("meses",
                                                                                             "anos",
                                                                                             "temp_min")
sea.set_context("poster")
f, ax = plt.subplots(figsize=(40, 10))
ax.set_title("Médias de temp. mínimas")
with sea.color_palette("Blues_r") as colors:
    sea.heatmap(temp_min, annot=True,linewidths=.5, ax=ax, cmap=colors)
```

```python
prec = pd.DataFrame({"meses":meses, "anos":anos, "prec":data.PrecipitacaoTotal}).pivot("meses",
                                                                                       "anos",
                                                                                       "prec")
sea.set_context("poster")
f, ax = plt.subplots(figsize=(130, 10))
ax.set_title("Médias de precipitação")
with sea.color_palette("RdBu_r") as colors:
    sea.heatmap(prec, annot=True,linewidths=.5, ax=ax, cmap=colors)
```
