```python
from bokeh.plotting import output_notebook, show, figure
import pandas as pd

output_notebook()
```

```python
nivel_oceanico = pd.read_csv("../../dados/nivel_oceanico.csv", header=5)
```

```python
nivel_oceanico.dtypes
```

```python
p = figure(plot_width=900, plot_height=300, title="Nvel ocenico")
c = (i for i in ["red", "orange", "yellow", "green", "blue", "purple"])
for i in nivel_oceanico.columns[1:]:
    p.scatter(nivel_oceanico.year, nivel_oceanico[i], color=next(c), legend=i)
p.legend.orientation="horizontal"
p.legend.location = (20, 250)
show(p)
```
